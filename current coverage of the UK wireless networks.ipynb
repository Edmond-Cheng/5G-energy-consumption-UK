```python
# use folium and pandas to visulize the current deployment of mobile networks
import folium 
import pandas as pd 
```

```python
station_data = pd.read_csv('basestation_sitefinder(2012).csv')
station_data.head()
```

```python
# instantiate a feature group for the incidents in the dataframe
basestation = folium.map.FeatureGroup()



# # loop through the all basestations and add each to the basestation feature group
for lat, lng in zip(station_data.Sitelat, station_data.Sitelng):
    basestation.add_child(
        folium.CircleMarker(
            [lat, lng],
            #swap
            radius=4, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=10
        )
    )
    


# # add basestation to map
uk_map = folium.Map(location=[51.38808, -2.33365], zoom_start=6)
uk_map.add_child(basestation)
```

```python
# labels of each base station
# add pop-up text to each marker on the map


for lat, lng, label in zip(station_data.Sitelat, station_data.Sitelng, station_data.Operator, 
                           station_data.Freqband, station_data.Technology):
    folium.Marker([lat, lng], popup=label).add_to(uk_map)    
    
# add incidents to map
uk_map.add_child(basestation)
```

```python
# heatmap
from folium.plugins import HeatMap

# let's start again with a clean copy of the map of San Francisco
san_map = folium.Map(location = [51.38808, -2.33365], zoom_start = 6)

# Convert data format
heatdata = station_data[['Sitelat','Sitelng']].values.tolist()

# add incidents to map
HeatMap(heatdata, radius = 20, min_opacity = 0.1, max_val = 50,gradient={.1: 'blue', .2: 'lime', .4: 'red'}).add_to(san_map)

san_map
```
